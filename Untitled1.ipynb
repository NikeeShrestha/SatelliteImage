```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
df=pd.read_csv("Finalprediction.csv")
df.columns
```

```python
plt.scatter(df.label, df.prediction)
plt.xlabel('RealValue')
plt.ylabel('PredictedValue')
r,pval=pearsonr(df['label'], df['prediction'])
plt.annotate('r = {:.2f}'.format(r), (min(df['label']), max(df['prediction'])), size=12)
plt.savefig("Validationvsreal.png")
```
